```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/19BCS2449/Time_series_rf/main/PS_Actuals.csv'
df = pd.read_csv(url, parse_dates=['Period'])
df.head()
```

```python
df.dropna(axis=0, inplace=True)
df.info()
```

```python
df = df[['Period', 'Currency', 'CostElement', 'BWN', 'Business Unit', 'Entity', 'Value']]
df.iloc[[0, -1]]
```

```python
df.sort_values(['Period'])
df.reset_index()
```

```python
Currency_list = df['Currency'].unique()
BWN_list = df['BWN'].unique()
CostElement_list = df['CostElement'].unique()
Business_list = df['Business Unit'].unique()
Entity_list = df['Entity'].unique()
print(f'Currency List:{Currency_list}')
print(f'BWN List:{BWN_list}')
print(f'Cost Element List:{CostElement_list}')
print(f'Business Unit List:{Business_list}')
print(f'Entity List:{Entity_list}')
```
